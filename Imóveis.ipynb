```python
#criando uma função

def pre(df,dataset_de_treino = True, cat_encoder = None ,std_scaler = None):

  dff = df.copy()

  #criando novas variaveis
  dff['comodos_por_familia']=dff['total_comodos']/dff['familias']
  dff['quartos_por_comodo']=dff['total_quartos']/dff['total_comodos']
  dff['populacao_por_familia']=dff['populacao']/dff['familias']

  #retirando os dados faltantes
  dff = dff.dropna(axis=0)

  normalizar = ['idade_mediana_das_casas','total_comodos','total_quartos',
              'populacao','familias','salario_mediano','comodos_por_familia',
              'quartos_por_comodo','populacao_por_familia']
  if dataset_de_treino:

    #dados quentes
    categorico = OneHotEncoder()
    df_prox_mar_OHE = categorico.fit_transform(dff[['proximidade_ao_mar']]).toarray()

    #nomalização
    sc = StandardScaler()
    normalizar = sc.fit_transform(dff[normalizar])

    X,y = np.c_[df_prox_mar_OHE,normalizar], dff.preco_mediano_das_casas.values
    return X, y,categorico,sc

  else:
      df_prox_mar_OHE = cat_encoder.transform(dff[['proximidade_ao_mar']]).toarray()

      normalizar = std_scaler.transform(dff[normalizar])

      X, y = np.c_[df_prox_mar_OHE,normalizar], dff.preco_mediano_das_casas.values
      return X, y










```

```python
Xtrain,ytrain ,categorico_train, scaler_train = pre(df = dftrain,
                                                   dataset_de_treino = True,
                                                   cat_encoder = None,
                                                   std_scaler = None)
```

```python
Xtrain.shape, ytrain.shape ,dftrain.shape
```

```python
Xtest, ytest = pre(df = dftest,
                   dataset_de_treino = False,
                   cat_encoder = categorico_train,
                   std_scaler = scaler_train)
```

```python
Xtest.shape,ytest.shape,dftest.shape
```

```python
#Importando os algoritmos
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
#metricas de performance
from sklearn.metrics import mean_squared_error, r2_score
```

```python
def rmse(ytrue, ypred):
  return np.sqrt(mean_squared_error(y_true = ytrue, y_pred = ypred))
```

```python
linear = LinearRegression()
linear.fit(Xtrain, ytrain)
```

```python
rmse(ytrue = ytrain, ypred = linear.predict(Xtrain))
```

```python
r2_score(y_true = ytrain, y_pred = linear.predict(Xtrain))
```

```python
vizinhos = [1, 2, 5, 10, 20, 50]

for n in vizinhos:
  knn = KNeighborsRegressor(n_neighbors = n)
  knn.fit(Xtrain, ytrain)

  preds = knn.predict(Xtrain)

  rmse_train = rmse(ytrue = ytrain, ypred = preds)
  r2_train = r2_score(y_true = ytrain,y_pred = preds)

  print('K = ' + str(n) + ' |rmse:' + str(np.round(rmse_train ,2)) +'/r2:'+ str(np.round(r2_train, 2)))
```

```python
#Arvores de regressão
max_profundidade = [1, 5, 10, 100]

for n in max_profundidade:
  dt = DecisionTreeRegressor(max_depth = n)
  dt.fit(Xtrain,ytrain)

  preds = dt.predict(Xtrain)

  rmse_train = rmse(ytrue = ytrain, ypred = preds)
  r2_train = r2_score(y_true = ytrain , y_pred = preds)

  print("Profundidade da arvore = " + str(n) + '|rmse:' + str(np.round(rmse_train,2)) + '/r2:' +str(np.round(r2_train,2)))
```

```python
# aplcar uma tecnica mais robustas, usando a validação cruzada

from sklearn.model_selection import cross_val_score
```

```python
CV = cross_val_score(estimator = LinearRegression(),
                     X = Xtrain,
                     y = ytrain,
                     cv = 5,
                     scoring = 'r2')

print(CV)
print('média:', CV.mean())
print('desvio padrão', CV.std())
```

```python
def cross(modelo, nome_modelo, X,y,num_folds):
  cv_resultados = cross_val_score(estimator = modelo,
                                  X = Xtrain,
                                  y = ytrain,
                                  cv = num_folds,
                                  scoring = 'r2')

  modelo.fit(Xtrain, ytrain)
  r2_train = r2_score(y_true = ytrain ,y_pred = modelo.predict(Xtrain))
  plt.figure(figsize = [8, 6])
  plt.scatter(range(1, 1 + num_folds), cv_resultados, c= 'blue', s = 100)
  plt.plot(range(1 ,1 + num_folds), cv_resultados , c='Blue', lw = 2 , label= 'R2 fold')
  plt.xticks(range(1, 1+ num_folds))
  plt.yticks(np.arange(0,1.1,0.1))
  plt.plot(range(1, 1+ num_folds),[cv_resultados.mean()]* num_folds, c= 'red', lw = 2,label = "R2 Cross-val Médio")

  plt.plot(range(1 , 1+num_folds),[cv_resultados.mean() + cv_resultados.std()]* num_folds, '--',
           c='gray',lw = 2, label = '+1 desvio padrão margem')
  plt.plot(range(1 , 1+num_folds),[cv_resultados.mean() - cv_resultados.std()]* num_folds, '--',
           c='gray',lw = 2, label = '-1 desvio padrão margem')

  plt.plot(range(1 ,1+num_folds), [r2_train] * num_folds, '--', c= 'black',lw = 3, label = 'r2 treino')
  plt.legend(fontsize = 10 ,bbox_to_anchor = [1.4,1.])
  plt.ylim([0,1.05])
  plt.grid()
  plt.ylabel('r2 score', fontsize = 14 ,fontweight = 'bold')
  plt.ylabel('Folds', fontsize = 14 , fontweight = 'bold')
  plt.title(nome_modelo + '| R2 Treino\valor:'+ str(np.round(r2_train,2)) + '/' + str(np.round(cv_resultados.mean(),2)),
            fontsize = 14 , fontweight = 'bold')
  plt.show()
```

```python
cross(modelo = LinearRegression(),
      nome_modelo = "Regressão Linear",
      X = Xtrain,
      y = ytrain,
      num_folds = 5)
```

```python
cross(modelo = KNeighborsRegressor(n_neighbors=10),
      nome_modelo = "knn1",
      X = Xtrain,
      y = ytrain,
      num_folds = 5)
```

```python
cross(modelo = DecisionTreeRegressor(max_depth=1),
      nome_modelo = "dectree(max_depth=1)",
      X = Xtrain,
      y = ytrain,
      num_folds = 5)
```
